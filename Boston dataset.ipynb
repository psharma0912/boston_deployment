### Independent and dependent features

```python
X=boston.drop('price', axis=1).values
y=boston['price'].values
```

### Train test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Standardising the dataset

```python
from sklearn.preprocessing import StandardScaler
Scaler=StandardScaler()
```

```python
X_train=Scaler.fit_transform(X_train)
X_test=Scaler.transform(X_test)
```

```python
import pickle
pickle.dump(Scaler, open('scaling.pkl','wb'))
```

### Model training

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

#### For linear regression model it is important to calculate intercept and slope

```python
#slope
print(regression.coef_)
```

```python
#Intercept
print(regression.intercept_)
```

```python
#On which parameters the model has been trained
print(regression.get_params())
```

### Prediction with the test data

```python
y_pred=regression.predict(X_test)
y_pred
```

```python
plt.scatter(y_test,y_pred)
#Plot a scatter plot for the prediction data
```

```python
#Residualplot
residual = y_test-y_pred

#PLotting the histogram of errors (histogram of residuals tells us about the shape of distribution of residuals)
#The graph follows a normal distribution although there seems to be a small number of outliers
sns.distplot(residual, bins=15)
plt.title('Error Terms')
plt.xlabel('Residual Errors')
plt.ylabel('freq')
```

```python
#Scatter plot of prediction wrt residuals
plt.scatter(y_pred,residual)
#There is a uniform distribution of residuals
```

### Evaluation metrics

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(np.sqrt(mean_squared_error(y_test,y_pred)))
      
```

```python
#R square and adjusted square
from sklearn.metrics import r2_score
n=len(X)
k=len(boston.columns)
k
```

```python
print('R^2_test:', r2_score(y_test, y_pred))

print('Adj(R2_test):', (1-((1-r2_score(y_test, y_pred))*(n-1)/(n-k-1))))
```
